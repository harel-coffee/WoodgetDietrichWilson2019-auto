## Combine datasets
Combines the 2016 and 2017 data to produce a dataset covering both years, which can then be used as an input to the ML process.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from ErrorML.ErrorML import *
```

```python
%matplotlib inline
```

```python
df_2016 = load_data('2016_ValidationPts_ALL_Updated17May2019_ROBIN.csv')
df_2017 = load_data('2017_ValidationPts_ALL_Update17May2019_ROBIN.csv')
#df_2017 = df_2017.drop(1516)
df_both = pd.concat([df_2016, df_2017])
df_both.to_csv('2016-17_ValidationPts_ALL_Updated17May2019_ROBIN_Combined.csv', index=False)
```

```python
df_2016['year'] = 2016
df_2017['year'] = 2017
df_both = pd.concat([df_2016, df_2017])
```
